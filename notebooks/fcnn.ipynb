```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
import numpy as np
```

```python
df = pd.read_csv(r"../data/airlines_label_encoding.csv", index_col=0)
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(df.drop(columns=["Delay"], axis=1))
y = df["Delay"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.value_counts()
```

```python
ros = RandomOverSampler(random_state=42)
X_train, y_train = ros.fit_resample(X_train, y_train)
```

```python
y_train.value_counts()
```

```python
X_columns = df.drop(columns=["Delay"], axis=1).columns
```

```python
X_df = pd.DataFrame(X, columns=X_columns)
X_df.head()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
```

```python
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(7,)),
    layers.BatchNormalization(),
    layers.Dense(64, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(1, activation='sigmoid')                     
])

model.compile(optimizer="adam",
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {accuracy:.4f}")
```

```python
model.save("fcnn.keras")
```
